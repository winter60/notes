```python
# question3  乱造的数据
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import rand


fig, ax = plt.subplots()
data = [
    ('#84badb', 'Cairns', np.arange(0.0, 50.0, 2.0)),
    ('#ffbd7d', 'Innisfail',np.arange(20, 70, 2.0))
]
for color,ltitle,x in data:
    y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
    ax.scatter(x, y, c=color, label=ltitle,
               alpha=0.8, edgecolors='none')

plt.xlabel("Length Fork")
plt.ylabel("Weight Total")
plt.legend(loc='upper left')
plt.show()
```

```python
# question4
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Letrhinus', 'Others', 'Epinepheius', 'Lutjanus'
sizes = [22.5,50.2,10.4,16.9]
explode = (.3, 0.4, .3, .5) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')


plt.show()
```
